```python
import pandas as pd
from  sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset = pd.read_csv('/content/Tiktok data.csv')
dataset = dataset.head(5000)
dataset
```

```python
dataset = dataset.drop(['claim_status', 'verified_status', 'video_share_count', 'video_download_count', 'video_comment_count'], axis=1)
```

```python
dataset
```

```python
dataset = dataset.dropna()
```

```python
dataset.duplicated().sum()
```

```python
import ast
```

```python
dataset['video_transcription_text'] = dataset['video_transcription_text'].apply(lambda x:x.split())
```

```python
dataset
```

```python
def drop_first_5_words(transcription):
    return transcription[6:]

df = dataset

df['video_transcription_text'] = df['video_transcription_text'].apply(drop_first_5_words)
```

```python
df
```

```python
df.info()
```

```python
df["video_transcription_text"] = df["video_transcription_text"].apply(lambda x:" ".join(x))
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['video_transcription_text']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps =  PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['video_transcription_text'] = df['video_transcription_text'].apply(stem)
df
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommendation(video_id):
    # Find the index of the video in the DataFrame based on video_id
    movi_index = df[df['video_id'] == video_id].index[0]

    # Get the similarity distances for the given video
    distance = similarity[movi_index]

    # Get the index of the top 10 most similar videos (the video itself is excluded)
    movi_list = sorted(enumerate(distance), reverse=True, key=lambda x: x[1])[1:11]

    # Print the video_ids of the top 10 most similar videos (excluding the provided video_id)
    for i in movi_list:
        # Retrieve the video_id from the DataFrame using iloc
        movie_video_id = df.iloc[i[0]]['video_id']
        print(movie_video_id)
```

```python
recommendation(9945378934) #Enter the User ID here
```
