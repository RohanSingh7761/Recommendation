```python
import pandas as pd
```

```python
from  sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder
```

```python
from datetime import datetime
```

```python
dataset = pd.read_csv('/UserData.csv')
dataset = dataset.head()
dataset
```

```python
interests = dataset['Interests'].str.get_dummies(', ') 
interests.fillna(0, inplace=True) 
interests
```

```python
gender = dataset[['Gender']]
gender_encoded = pd.get_dummies(gender)
gender_encoded
```

```python
location = dataset[['City', 'Country']]
location_encoded = pd.get_dummies(location)
location_encoded = location_encoded.astype(int)
location_encoded
```

```python
features = pd.concat([interests, gender_encoded, location_encoded], axis=1)
features
```

```python
similarity_matrix = cosine_similarity(features)
```

```python
user_id = 110 #Enter user ID here
similar_users_indices = similarity_matrix[user_id - 1].argsort()[::-1]  #This is to sort in descending order
similar_users = similar_users_indices[1:11]  # To get top 10 user
```

```python
for similar_user_index in similar_users:
    print(f"Similar User ID: {similar_user_index + 1}")
```
